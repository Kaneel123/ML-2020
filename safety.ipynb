```python
#import the libraries 
import numpy as np


# In[169]:


#create a function to do the flipping

def flip(state, flip_loc, algorithm):
    """
    variables
    ---------
    state - the current state of the pancake
    flip_loc - location of the flip. This should each character. So it should be either (2,4,6,8)
    algorithm - what is the algorithm used
    
    output
    ---------
    next_state - flips the current state given the flip_location
    print_flip - shows where the flip happens in the current state using a '|'
    flip_loc - the flip location (2,4,6,8)
    """
    if algorithm == 'b':
        
        print_flip = state[:flip_loc] + '|' + state[flip_loc:]
        
        if flip_loc == 2:
            #if flip location is 2
            state_flip = state[:flip_loc]

            if state_flip[flip_loc - 1] == 'w':
                state_flip = state_flip.replace('w','b')
            else:
                state_flip = state_flip.replace('b','w')

        if flip_loc == 4:
            state_flip = state[:flip_loc]

            #flip the two numbers
            #put the state_flip to a list
            state_flip_list = list(state_flip)

            #change the order of the first character and number
            state_flip_list[0] = state_flip[2] #number
            state_flip_list[1] = state_flip[3] #character

            #change the order of the second character and number        
            state_flip_list[2] = state_flip[0]
            state_flip_list[3] = state_flip[1]

            #now change the character (if b change to w, if w change to b)

            for idx in [1,3]:           

                if state_flip_list[idx] == 'w':
                    state_flip_list[idx] = 'b'                               
                else:
                    state_flip_list[idx] = 'w'

            #change the state flip list to a string
            state_flip = ''.join(state_flip_list)


        if flip_loc == 6:
            state_flip = state[:flip_loc]

            #flip the numbers (followign the same procedure as above)
            state_flip_list = list(state_flip)

            #change the numbers of the first and the third and keep the middle one.
            state_flip_list[0] = state_flip[4]
            state_flip_list[1] = state_flip[5]

            state_flip_list[4] = state_flip[0]
            state_flip_list[5] = state_flip[1]

            #now change w's to b's and b's to w's
            for idx in [1,3,5]:
                if state_flip_list[idx] == 'w':
                    state_flip_list[idx] = 'b'
                else:
                    state_flip_list[idx] = 'w'

            state_flip = ''.join(state_flip_list)

        if flip_loc == 8:
            state_flip = state[:flip_loc]

            #flip the numbers(same as above but here switching the 4th and 1st pair and then 3rd and 2nd)
            state_flip_list = list(state_flip)
            state_flip_list[0] = state_flip[6] #fourth to first number
            state_flip_list[1] = state_flip[7] #fourth to first character
            state_flip_list[6] = state_flip[0] #first to fourth number
            state_flip_list[7] = state_flip[1] #first to fourth character

            state_flip_list[4] = state_flip[2] #3rd to 2nd number
            state_flip_list[5] = state_flip[3] #3rd to 2nd character
            state_flip_list[2] = state_flip[4] #2nd to 3rd number
            state_flip_list[3] = state_flip[5] #2nd to 3rd character

            #now change w's to b's and b's to w's
            for idx in [1,3,5,7]:
                if state_flip_list[idx] == 'w':
                    state_flip_list[idx] = 'b'
                else:
                    state_flip_list[idx] = 'w'

            state_flip = ''.join(state_flip_list)
          
    elif algorithm == 'a':
        
        print('have not done it yet')
        
    next_state = state_flip + state[flip_loc:]
        
    return next_state, print_flip, flip_loc


def expand(flip_loc_prev):
    """
    variables
    ---------
    flip_loc_prev - takes the previous flip location
    
    output
    --------
    flip - flips available without going back to the previous state    
    """
    
    if flip_loc_prev == 2:
        flip = 4, 6, 8
            
    elif flip_loc_prev == 4:
        flip = 2, 6, 8
    
    elif flip_loc_prev == 6:
        #flip = 2, 4, 8
        flip = 4, 2, 8
        
    elif flip_loc_prev == 8:
        #flip = 2, 4, 6
        flip = 4, 2, 6
        
    return flip 

def bfs(initial_state, algorithm):
    
    # create empty lists to append all the needed flips and states
    q = [] #this is the cue
    nodes = [] #this append the flips
    qq = []
    
    # define the initial state and the goal state. The initial state is an input to the function
    initial_state = initial_state
    goal_state = '1w2w3w4w'
    
    #create a dictionaries to append all the states
    graph = {}
    graph_nodes = {}
    g_dict = {}
    
    for i in [2,4,6,8]:
        #this loop appends the first four
        next_state, print_flip, current_flip  = flip(initial_state, i, 'b')
        
        q.append(next_state) #append the first four states to the qeue
        qq.append(next_state)
        nodes.append(current_flip) #append the first four nodes to the a different qeue
        g_dict[current_flip] = next_state
    
        
    graph[initial_state] = q
    graph_nodes[initial_state] = g_dict
    
    
    #run a while loop to create the graph
    search = True
    while search == True:
        
        #expand the first node
        expand_nodes = expand(nodes[0])
        
        #create an empty list and a dictionary 
        g = []
        g_dict = {}
        
        for i in range(len(expand_nodes)):
            
            next_state, _, current_flip = flip(q[0], expand_nodes[i], algorithm)
            
            if next_state not in qq:
                q.append(next_state)
                qq.append(next_state)
                nodes.append(current_flip)
                g.append(next_state)
                g_dict[current_flip] = next_state
        
        
        #store everything 
        graph[q[0]] = g
        graph_nodes[q[0]] = g_dict
        
        if q[0] == goal_state:
            search = False
            
        q.pop(0)
        nodes.pop(0)      
    
    
    
    return graph, graph_nodes


# In[219]:


def print_result(graph_nodes, algorithm):   


    parent = []
    sub_nodes = []

    for key, value in graph_nodes.items():
        parent.append(key)
        sub_nodes.append(value)

    state = '1w2w3w4w'

    opt_path = []
    opt_path_node = []

    search = True

    while search == True:

        for i in range(len(sub_nodes)):
            
            if state in sub_nodes[0].values():
                search = False
            
            elif state in sub_nodes[i].values():
                for key, value in sub_nodes[i].items():
                    if value == state:
                        opt_path_node.append(key)

                opt_path.append(parent[i])
                state = parent[i]

            
                
    
    opt_path.append(parent[0])         
    
    for k,v in sub_nodes[0].items():
        if v == state:
            opt_path_node.append(k) 
            
              
     
    #print result

    for i in range(len(opt_path)):
        _,print_flip,_ = flip(np.flip(opt_path, axis = 0)[i], np.flip(opt_path_node, axis = 0)[i], algorithm)
        print(print_flip)
    print('1w2w3w4w')


# In[220]:



user_input = input()

initial_state = user_input[:8]
algorithm = user_input[9]

_, graph_node = bfs(initial_state, algorithm)

print_result(graph_node,algorithm)
```
